```python
industry = {'건축물':0, '화학 / 석유화학':1, '철강':2, '전기전자 (반도체 디스플레이 포함)':3, '식품':4, '수송기계':5, '기계':6, '제조업 기타':7,
            '요업':8, '섬유/ 피혁':9, '제지목재':10, '폐기물':11, '정유':12, '발전 / 에너지 - 수도':13, '시멘트':14, '비철금속':15, '발전 / 에너지 - 가스':16,
            '-':17, '광업':18, '농업, 임업 및 어업':19, '수송':20}
```

```python
df['INDUTY'].map(industry)
```

```python
df['industry']=df['INDUTY'].map(industry)
```

```python
df.head()
```

```python
df_feature=df[['INDUTY','industry','TRGT_EQMT','CLSF_IMPRV','IMPRV_ACT','AMNT_RDC_GHG']]
```

```python
df_feature.head()
```

```python
# CO2 감축량을 기준으로 내림차순 정렬
df_sort_top5=df_feature.sort_values(by='AMNT_RDC_GHG',ascending=False).groupby('industry').head(5)
df_sort_top5
```

```python
df_sort_top5.sort_values(by=['industry','AMNT_RDC_GHG'],ascending=[True,False])
```

```python
def top(df_feature, n=5, column='AMNT_RDC_GHG'):
  return df_feature.sort_values(by='AMNT_RDC_GHG',ascending=False)[:n]

df_feature.groupby('industry').apply(top, column='industry',n=5)
```
