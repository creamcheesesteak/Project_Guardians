## IMPRV_ACT 컬럼(개선활동명) 텍스트 분석 하다 망한것

```python
text01 = 'Filtrated Brine Pump(PU-1154A/S)펌프 유량 및 양정 최적화 '
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
def remove_punctuation(text):
  tokenizer = RegexpTokenizer('[(가-힣a-zA-Z0-9-/)]+')
  text_list = tokenizer.tokenize(text)
  result = ' '.join(text_list)
  return result
```

```python
remove_punctuation(text01)
```

```python
df_tech['cleaned'] = df_tech['IMPRV_ACT'].apply(remove_punctuation)
```

```python
df_tech.info()
```

```python
!curl -O https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz
```

```python
from gensim.models import Word2Vec
```

```python
corpus = []
for words in df_tech['cleaned']:
  corpus.append(words.split())
print(type(corpus),corpus[4:8])
```

```python
 word2vec_model = Word2Vec(min_count=2, workers=-1, size=300)
 word2vec_model.build_vocab(corpus) # need to modif
```

```python
word2vec_model.intersect_word2vec_format('GoogleNews-vectors-negative300.bin.gz', lockf=1.0, binary=True)
```

```python
word2vec_model.train(df_tech['cleaned'],total_examples=word2vec_model.corpus_count, epochs=15)
```

```python
word2vec_model.wv.vocab
```

```python
doc_embedding_list = list()
# for line in df_data['cleaned'][2:3]:
for line in df_tech['cleaned']:
  doc2vec = None
  for word in line.split():
    if word in word2vec_model.wv.vocab:
      if doc2vec is None:
        doc2vec = word2vec_model[word] 
      else :
        doc2vec = doc2vec + word2vec_model[word] 

  if doc2vec is not None:
    doc2vec = doc2vec / len(doc2vec)
    doc_embedding_list.append(doc2vec)

print(doc_embedding_list)
```

```python
type(doc_embedding_list), doc_embedding_list[3]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(doc_embedding_list, doc_embedding_list)
type(cosine_sim), cosine_sim.shape
```

## 새로 하는것

```python
import pandas as pd
```

```python
df_tech_0 = pd.read_csv('(0이하_값_전처리_후)에너지_및_온실가스_감축기술.csv',encoding='cp949')
df_tech_0  = df_tech_0.drop('Unnamed: 0', axis=1)
```

```python
df_tech_0.info()
```

```python
dfx_INDUTY = pd.get_dummies(df_tech_0['INDUTY'])
dfx_INDUTY
```

```python
dfx_TRGT_EQMT = pd.get_dummies(df_tech_0['TRGT_EQMT'])
dfx_TRGT_EQMT
```

```python
dfx_CLSF_IMPRV = pd.get_dummies(df_tech_0['CLSF_IMPRV'])
dfx_CLSF_IMPRV
```

```python
from sklearn import cluster
kmeans1 = cluster.KMeans(n_clusters=20)
kmeans2 = cluster.KMeans(n_clusters=22)
kmeans3 = cluster.KMeans(n_clusters=90)
```

```python
kmeans1.fit(dfx_INDUTY)
kmeans1.labels_
```

```python
kmeans2.fit(dfx_TRGT_EQMT)
kmeans2.labels_
```

```python
kmeans3.fit(dfx_CLSF_IMPRV)
kmeans3.labels_
```

```python
df_tech_0['INDUTY_label'] = kmeans1.labels_
df_tech_0['TRGT_EQMT_label'] = kmeans2.labels_
df_tech_0['CLSF_IMPRV_label'] = kmeans3.labels_
```

```python
df_tech_0.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = 'INDUTY_label',y = 'TRGT_EQMT_label',data= df_tech_0,fit_reg=False,size=6, hue='CLSF_IMPRV_label')
plt.show()
```

```python
sns.lmplot(x = 'TRGT_EQMT_label',y = 'CLSF_IMPRV_label',data= df_tech_0,fit_reg=False,size=6, hue='INDUTY_label')
plt.show()
```

```python
sns.lmplot(x = 'TRGT_EQMT_label',y = 'CLSF_IMPRV_label',data= df_tech_0,fit_reg=False,size=6, hue='INDUTY_label')
plt.show()
```
