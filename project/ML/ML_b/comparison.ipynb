```python
import pandas as pd
```

```python
df4 = pd.read_csv('에너지_및_온실가스_감축기술.csv')
```

```python
df4.head(10)
```

```python
df4.columns = ['연번','업종','감축 대상 설비','개선 구분','개선활동명','투자비 회수기간','투자비 회수기간 단위','절감액','절감액 단위','투자비','투자비 단위','진단 연도','에너지 사용량 규모','에너지 절감 종류','에너지 절감량(연료)','연료 단위','에너지 절감량(전력)','전력 단위','온실가스 감축량','감축량 단위']
```

```python
df4.head(10)
```

```python
df5 = df4.loc[:,['업종','감축 대상 설비','개선 구분','개선활동명','에너지 사용량 규모','에너지 절감 종류','온실가스 감축량']]
```

```python
df5.head(10)
```

```python
df5['에너지 절감 종류'].unique()
```

```python
df5['에너지 사용량 규모'].unique()
```

```python
df_INDUTY = df5[df5['업종']=='-'].index
df_TRGT_EQMT = df5[df5['감축 대상 설비']=='-'].index
df_CLSF_IMPRV = df5[df5['개선 구분']=='-'].index
df_IMPRV_ACT = df5[df5['개선활동명']=='-'].index
df_ENG_CNSM_SCL = df5[df5['에너지 사용량 규모']=='-'].index
```

```python
df6 = df5.drop(df_INDUTY)
df6 = df6.drop(df_TRGT_EQMT)
df6 = df6.drop(df_CLSF_IMPRV)
df6 = df6.drop(df_IMPRV_ACT)
df6 = df6.drop(df_ENG_CNSM_SCL)
```

```python
df6.info()
```

```python
df6['에너지 사용량 규모'].unique()
```

```python
df5[df5['에너지 절감 종류'].str.contains('2011')]
```

```python
df5[df5['에너지 절감 종류'].str.contains('2017')]
```

```python
df6.head(2)
```

```python
df_에너지_사용량_규모 = pd.get_dummies(df6['에너지 사용량 규모'])
df_에너지_사용량_규모
```

```python
df_업종 = pd.get_dummies(df6['업종'])
df_업종
```

```python
df_개선_구분 = pd.get_dummies(df6['개선 구분'])
df_개선_구분
```

```python
from sklearn import cluster
kmeans1 = cluster.KMeans(n_clusters=7)
kmeans2 = cluster.KMeans(n_clusters=20)
kmeans3 = cluster.KMeans(n_clusters=90)
```

```python
kmeans1.fit(df_에너지_사용량_규모)
kmeans1.labels_

kmeans2.fit(df_업종)
kmeans2.labels_

kmeans3.fit(df_개선_구분)
kmeans3.labels_
```

```python
df6['에너지_원핫'] = kmeans1.labels_
df6['업종_원핫'] = kmeans2.labels_
df6['개선구분_원핫'] = kmeans3.labels_
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.lmplot(x = '에너지_원핫',y = '업종_원핫',data=  df6,fit_reg=False,size=6, hue='개선구분_원핫')
plt.yticks(np.arange(0, 20))
plt.show()
```
