```python
import pandas as pd
```

```python
df = pd.read_excel('2019.xlsx')
```

```python
top = df['특성추출(가중치순 상위 50개)']
```

```python
words = []

for i in range(len(top)):
    a = str(top[i])
    wordlist = a.split(',')
    words.append(wordlist)   
```

```python
word_list = []
for word in words:
    word_list += word
```

```python
wordcount = {}

for word in word_list:
    wordcount[word] = wordcount.get(word, 0) + 1
    keys = sorted(wordcount.keys())
```

```python
df = pd.DataFrame(wordcount, index = ['count'])
df = df.transpose()
df['year'] = 2019
df
```
